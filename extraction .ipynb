```python
import pandas as pd
from ntscraper import Nitter
import pymongo

scraper = Nitter()
```

```python
def get_tweets(hashtag, mode, size):
    tweets = scraper.get_tweets(hashtag, mode=mode, number=size)
    final_tweets = []

    for tweet in tweets['tweets']:
        data_tweet = [tweet['link'], tweet['text'], tweet['date'], tweet['stats']['likes'], tweet['stats']['comments']]
        final_tweets.append(data_tweet)

    data_df = pd.DataFrame(final_tweets, columns=['link', 'text', 'date', 'Likes', 'Comments'])
    return data_df
```

```python
data_tweets = get_tweets("Openai", "hashtag", 15)
```

```python
client = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = client['tweets']
tweetsinfo = mydb['tweets_palestine']  # Renommez la collection
for index, row in data_tweets.iterrows():
    tweet_dict = row.to_dict()
    tweetsinfo.insert_one(tweet_dict)
```
